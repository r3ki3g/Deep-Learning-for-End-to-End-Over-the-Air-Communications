```python
import json
import os
import sys


from matplotlib import pyplot as plt
```

```python
archi_res_dir = r"D:\ENTC\PROJECTS\Deep-Learning-for-End-to-End-Over-the-Air-Communications\Sequence Decoder\archi_tests"
```

```python
all_experiment_rec_files = os.listdir(archi_res_dir)
all_experiment_rec_files[0]
```

```python
all_experiments = []

for rec_file in all_experiment_rec_files:
        with open(os.path.join(archi_res_dir,rec_file),"r") as file:
            experiment_obj = json.load(file)
            all_experiments.append(experiment_obj)

    
        
        
```

```python
all_experiments[0]
```

```python
def plot_experiments_loss(experiment_obj):
    J_list = experiment_obj['J_list']
    
    title = "n_front_dense_FE: {}, \
    n_front_dense_PE: {}, n_front_dense_RxDec: {}, \
    state_width: {}, n_epochs: {}".format(experiment_obj['n_front_dense_FE'],
                                          experiment_obj['n_front_dense_PE'],
                                          experiment_obj['n_front_dense_RxDec'],
                                          experiment_obj['state_width'],
                                         experiment_obj['n_epochs'])
    plt.plot(J_list)
    plt.title(title)
    plt.show()
```

```python
plot_experiments_loss(all_experiments[50])
```
